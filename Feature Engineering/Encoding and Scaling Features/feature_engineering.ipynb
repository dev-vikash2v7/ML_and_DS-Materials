# 1. Min Max Normalizing

```python
#min max normalise

from sklearn import preprocessing
data = {'rupaye' : [45 , 5 , 334, 23 , 43 , 56]}
rupaye_frame = pd.DataFrame(data)

min_max_normalizer = preprocessing.MinMaxScaler()
scaled_data = min_max_normalizer.fit_transform(rupaye_frame)
#scaled_data = min_max_normalizer.transform(test_frame)
rupaye_frame_normalized = pd.DataFrame(scaled_data)
rupaye_frame_normalized
```

# bining numerical features


```python

data = {'age' : [45 , 5 , 34, 23 , 43 , 56 , 6 , 42, 7]}
age_frame = pd.DataFrame(data)

bins = [0 , 20 , 40   , 60  , 80]
group_name = [ 'less than 20 yeas ' , ' 20 - 40 years' , '40-60 years' , '60+']

age_frame['categories'] = pd.cut(age_frame['age'] , bins , labels = group_name )
age_frame
```

# 4. One hot encoding

```python
#creating dummy variable form categorical data
dummy_frame = pd.get_dummies(age_frame.categories)
dummy_frame
```

# creating word counter vector
transform a given text into a vector on the basis of the frequency (count) of each word that occurs in the entire text


```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
comments = ['nice product' , 'bad condition' , 'shipping was bad' , 'great delivery' ]
count_vect = CountVectorizer()

# Printing the identified Unique words along with their indices
X_train_counts = count_vect.fit(comments)
X_train_counts.vocabulary_

#word_frame = pd.DataFrame(X_train_counts.toarray())
#word_frame #Sparse Matrix.

vector = X_train_counts.transform(comments)
  
# Summarizing the Encoded Texts
print("Encoded Document is:")
vector = vector.toarray()
print(vector)
```
