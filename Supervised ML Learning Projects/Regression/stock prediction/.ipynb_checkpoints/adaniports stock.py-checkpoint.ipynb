# 1-step = setupping and cleaning the data

```python
stock = pd.read_csv('ADANIPORTS.csv' , index_col = 'Date' ,parse_dates = True)
```

# 2-step =  training the data

```python
training_set = stock['Open']
training_set = pd.DataFrame(training_set)
training_set.head(3)
```

```python
from sklearn.preprocessing import MinMaxScaler

#trans features  by scaling each fearure to a giveen range
#scales and translates each features individually such that it is int the given rnge on the trainniing set, eg; btw 0 - 1
scaler = MinMaxScaler(feature_range=(0,1))

#fit to data then transform it
#fit trans to X and Y with optional param fit_params and returns a trans version of X
training_set_scaled = scaler.fit_transform(training_set) 

#retrun transformed array
len(training_set_scaled) , training_set_scaled
```

```python
#creating a data str with 60 timesteps and 1 output
X_train = []
Y_train= []
for i in range(60 , 1258):
    X_train.append(training_set_scaled[i-60 : i ,0])
    Y_train.append(training_set_scaled[i , 0])
    
X_train = np.array(X_train)
Y_train = np.array(Y_train)

print(X_train)
print(Y_train)
len(X_train) ,len(Y_train)

X_train.shape , Y_train.shape
```

```python
#reshapping
X_train = np.reshape(X_train  , (X_train.shape[0] , X_train.shape[1] , 1))
X_train.shape
```
